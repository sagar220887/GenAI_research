```python
import pandas as pd

df = pd.read_csv('../data/census.csv')
df.head()
```

### Local LLM

```python
from pandasai.llm.local_llm import LocalLLM
from pandasai import SmartDataframe

ollama_llm = LocalLLM(api_base="http://localhost:11434/v1", model="codellama")


df = SmartDataframe(df, config={"llm": ollama_llm, "custom_whitelisted_dependencies": ["pandasai"]})

response = df.chat("What are the ctyname for state Alabama?")
```
